```python
import matplotlib.pyplot as plt

cardata['Kms_Driven'].plot(kind='hist', bins=20, edgecolor='k')
plt.xlabel('Kilometraje')
plt.ylabel('Frecuencia')
plt.title('Histograma del Kilometraje de Automóviles')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='Fuel_Type', y='Kms_Driven', data=cardata)
plt.xlabel('Tipo de Combustible')
plt.ylabel('Kilometraje')
plt.title('Diagrama de Caja del Kilometraje por Tipo de Combustible')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
cardata['Year'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Año')
plt.ylabel('Cantidad de Automóviles')
plt.title('Distribución de Automóviles por Año de Fabricación')
plt.show()
```
